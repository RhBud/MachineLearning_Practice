```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()

X = iris.data

Y = iris.target
```

```python
print iris.DESCR
#just a reminder of the documentation
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y)
```

```python
model.fit(X_train,Y_train)
```

```python
from sklearn import metrics
```

```python
predicted = model.predict(X_test)
```

```python
expected = Y_test
```

```python
print metrics.accuracy_score(expected,predicted)
```

```python
#now we're going to compare to four different methods
#two linear models, a Gaussian Radial Basis Function, and a SVC with a polynomial (3rd Degree) kernel
```

```python
#the two linear models are LinearSVC which uses the One-vs-all multiclass reduction while SVC uses the One-vs-One muliclass reduction
```

```python
from sklearn import svm
```

```python
#not going to use all the data, and not going to split it
```

```python
X = iris.data[:,:2]
#select first two features

Y = iris.target
```

```python
C = 1.0
# SVM regularization parameter
```

```python
svc = svm.SVC(kernel='linear',C=C).fit(X,Y)
```

```python
#now gaussian radial basis function (RBF)
rbf_svc = svm.SVC(kernel='rbf',gamma=0.7,C=C).fit(X,Y)
```

```python
poly_svc = svm.SVC(kernel='poly',degree=3,C=C).fit(X,Y)
```

```python
lin_svc = svm.LinearSVC(C=C).fit(X,Y)
```

```python
#now we have our four models
```

```python
#going to use a mesh grid now
```

```python
h = 0.02
#stepsize for meshgrid

x_min = X[:,0].min() -1
x_max = X[:,0].max() +1

y_min = X[:,1].min() -1
y_max = X[:,1].max() +1

```

```python
#create a meshgrid in numpy
xx,yy = np.meshgrid(np.arange(x_min,x_max,h),np.arange(y_min,y_max,h))
```

```python
titles = ['SVC with linear kernel',
          'LinearSVC (linear kernel)',
          'SVC with RBF kernel',
          'SVC with polynomial (degree 3) kernel']
```

```python
#now we are making a for loop to go through the models and to plot them
for i,clf in enumerate((svc,lin_svc,rbf_svc,poly_svc)):
    
    plt.figure(figsize=(15,15))
    
    plt.subplot(2,2,i+1)
    #depending on figuresize we might not be able to fit all the subplots, have to play with it. enumeration starts at 1 not 0
    
    plt.subplots_adjust(wspace=0.4,hspace =0.4)
    
    Z = clf.predict(np.c_[xx.ravel(),yy.ravel()])
    #have to pass a transformed version of the arrays using np
    
    Z = Z.reshape(xx.shape)
    
    plt.contourf(xx,yy,Z,cmap=plt.cm.terrain,alpha=0.5,linewidths=0)
    
    plt.scatter(X[:,0],X[:,1],c=Y,cmap=plt.cm.Dark2)
    
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.xlim(xx.min(),xx.max())
    plt.ylim(yy.min(),yy.max())
    
    plt.xticks(())
    plt.yticks(())
    plt.title(titles[i])
```
